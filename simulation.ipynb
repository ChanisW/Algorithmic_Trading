```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from agent_ma import MovingAverageCrossoverAgent
from agent_mm import MomentumAgent
from agent_rsi import RSIAgent
from agent_ema import EMAAgent

# Function to fetch live data from Binance API
def fetch_live_data():
    url = 'https://api.binance.com/api/v3/ticker/price?symbol=BTCUSDT'
    response = requests.get(url)
    data = response.json()
    return float(data['price'])

# Function to fetch historical data from Binance API
def fetch_historical_data(symbol, interval, limit=1000):
    url = f'https://api.binance.com/api/v3/klines?symbol={symbol}&interval={interval}&limit={limit}'
    res = requests.get(url)
    data = res.json()
    df = pd.DataFrame(data, columns=[
        'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 
        'Close_time', 'Quote_asset_volume', 'Number_of_trades', 
        'Taker_buy_base_asset_volume', 'Taker_buy_quote_asset_volume', 'Ignore'
    ])
    df["Date"] = pd.to_datetime(df["Date"], unit="ms")
    df[["Open", "High", "Low", "Close", "Volume"]] = df[["Open", "High", "Low", "Close", "Volume"]].astype(float)
    df = df[["Date", "Open", "High", "Low", "Close", "Volume"]]
    return df

# Preprocess data to add Returns column
def preprocess_data(df):
    df['Returns'] = df['Close'].pct_change()
    return df

# Update historical data by appending new data and forward filling
def update_historical_data(df_historical, new_data):
    new_data_filled = new_data.ffill()
    if not new_data_filled.empty:
        df_historical = pd.concat([df_historical, new_data_filled])
    return df_historical

# Train agents on historical data
def train_agents(agents, historical_data):
    for agent in agents:
        agent.train(historical_data)  # Assuming each agent has a train method
        print(f"{agent.name} trained.")

# Backtesting function for agents using historical data
def backtest_agents(agents, historical_data):
    df_historical = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])
    portfolio_values = {agent.name: [] for agent in agents}
    
    for index, row in historical_data.iterrows():
        new_data = pd.DataFrame([row])
        df_historical = update_historical_data(df_historical, new_data)
        
        for agent in agents:
            agent.trade(df_historical)
            portfolio_values[agent.name].append(agent.get_portfolio_value(row["Close"]))
    
    return portfolio_values

# Fetch historical data for backtesting for each interval
intervals = {
    "1m": "1m",
    "1h": "1h",
    "4h": "4h",
    "1d": "1d"
}

# Initialize agents
mac_agent = MovingAverageCrossoverAgent()
mm_agent = MomentumAgent()
rsi_agent = RSIAgent()
ema_agent = EMAAgent()
agents = [mac_agent, mm_agent, rsi_agent, ema_agent]

# Train the agents
print("Training agents...")
for interval_name, interval in intervals.items():
    historical_data = fetch_historical_data("BTCUSDT", interval)
    historical_data = preprocess_data(historical_data)
    train_agents(agents, historical_data)
    print(f"Agents trained for interval: {interval_name}")

# Backtest the agents for each interval and collect data for plotting
portfolio_values_all = {agent.name: {interval: [] for interval in intervals.keys()} for agent in agents}
dates_all = {interval: [] for interval in intervals.keys()}

for interval_name, interval in intervals.items():
    print(f"Backtesting for interval: {interval_name}")
    historical_data = fetch_historical_data("BTCUSDT", interval)
    historical_data = preprocess_data(historical_data)
    portfolio_values = backtest_agents(agents, historical_data)
    
    # Collect data for plotting
    for agent_name, values in portfolio_values.items():
        portfolio_values_all[agent_name][interval] = values
    dates_all[interval] = historical_data['Date']

# Plot portfolio values of all agents on one graph
plt.figure(figsize=(14, 7))

for agent_name, intervals_data in portfolio_values_all.items():
    for interval, values in intervals_data.items():
        plt.plot(dates_all[interval], values, label=f'{agent_name} ({interval})')

plt.title("Portfolio Values of All Agents Over Time")
plt.xlabel("Date")
plt.ylabel("Portfolio Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import requests
import pandas as pd
import time
from agent_ma import MovingAverageCrossoverAgent
from agent_mm import MomentumAgent
from agent_rsi import RSIAgent
from agent_bb import BollingerBandsAgent
from agent_ema import EMAAgent
from agent_my import MyOwnAgent

# Function to fetch live data from Binance API
def fetch_live_data():
    url = 'https://api.binance.com/api/v3/ticker/price?symbol=BTCUSDT'
    response = requests.get(url)
    data = response.json()
    return float(data['price'])

# Preprocess data to add Returns column
def preprocess_data(df):
    df['Returns'] = df['Close'].pct_change()
    return df

# Update live data with new price and preprocess
def update_data(df):
    price = fetch_live_data()
    timestamp = pd.Timestamp.now()
    new_data = pd.DataFrame({"Date": [timestamp], "Close": [price]})
    df = pd.concat([df, new_data], ignore_index=True)
    df.set_index("Date", inplace=True)
    df = preprocess_data(df)
    return df

# Update agent decisions based on new data
def update_agents(df, agents):
    df = update_data(df)
    current_price = df["Close"].iloc[-1]
    for agent in agents:
        agent.trade(df)
        print(f"{agent.name} Portfolio Value: {agent.get_portfolio_value(current_price)}")
    return df

# Function to run live trading with periodic updates
def run_live(df, interval, *agents):
    intervals = {
        "1m": 60,
        "1h": 3600,
        "4h": 14400,
        "1d": 86400
    }
    while True:
        print(f"{interval} update")
        df = update_agents(df, agents)
        print("="*40)
        time.sleep(intervals[interval])

# Fetch initial historical data
intervals = ["1m", "1h", "4h", "1d"]

# Initialize agents
mac_agent = MovingAverageCrossoverAgent()
mm_agent = MomentumAgent()
rsi_agent = RSIAgent()
ema_agent = EMAAgent()
agents = [mac_agent, mm_agent, rsi_agent, ema_agent]

# Run live trading simulation for each interval
for interval in intervals:
    print(f"Starting live trading simulation for interval: {interval}")
    df = fetch_historical_data("BTCUSDT", interval)
    df = preprocess_data(df)
    run_live(df, interval, *agents)
```

```python
import numpy as np
import pandas as pd

# Preprocess data to add Returns column
def preprocess_data(df):
    df['Returns'] = df['Close'].pct_change()
    return df

# Function to calculate total return
def calculate_total_return(portfolio_values):
    initial_value = portfolio_values[0]
    final_value = portfolio_values[-1]
    return (final_value - initial_value) / initial_value

# Function to calculate Sharpe ratio
def calculate_sharpe_ratio(returns, risk_free_rate=0):
    excess_returns = returns - risk_free_rate
    return np.mean(excess_returns) / np.std(excess_returns)

# Function to calculate maximum drawdown
def calculate_max_drawdown(portfolio_values):
    drawdowns = []
    peak = portfolio_values[0]
    for value in portfolio_values:
        if value > peak:
            peak = value
        drawdowns.append((peak - value) / peak)
    return max(drawdowns)

# Function to evaluate performance metrics
def evaluate_performance(portfolio_values):
    returns = np.diff(portfolio_values) / portfolio_values[:-1]
    total_return = calculate_total_return(portfolio_values)
    sharpe_ratio = calculate_sharpe_ratio(returns)
    max_drawdown = calculate_max_drawdown(portfolio_values)
    
    return {
        "Total Return": total_return,
        "Sharpe Ratio": sharpe_ratio,
        "Max Drawdown": max_drawdown
    }

# Function to convert performance metrics to DataFrame
def metrics_to_dataframe(performance_metrics):
    rows = []
    for agent_name, intervals in performance_metrics.items():
        for interval, metrics in intervals.items():
            row = {"Agent": agent_name, "Interval": interval}
            row.update(metrics)
            rows.append(row)
    return pd.DataFrame(rows)

# Initialize agents
mac_agent = MovingAverageCrossoverAgent()
mm_agent = MomentumAgent()
rsi_agent = RSIAgent()
ema_agent = EMAAgent()
agents = [mac_agent, mm_agent, rsi_agent, ema_agent]

# Define intervals for testing
intervals = {
    "1m": "1m",
    "1h": "1h",
    "4h": "4h",
    "1d": "1d"
}

# Initialize a dictionary to store performance metrics
performance_metrics = {agent.name: {interval: None for interval in intervals.keys()} for agent in agents}

# Backtest the agents for each interval and evaluate performance
for interval_name, interval in intervals.items():
    print(f"Running backtest for interval: {interval_name}")
    historical_data = fetch_historical_data("BTCUSDT", interval)
    historical_data = preprocess_data(historical_data)
    portfolio_values = backtest_agents(agents, historical_data)
    
    for agent_name, values in portfolio_values.items():
        print(f"Agent: {agent_name} - Interval: {interval_name}")
        
        # Evaluate performance metrics
        performance = evaluate_performance(values)
        performance_metrics[agent_name][interval_name] = performance
        
        print(f"Performance for {agent_name} - Interval: {interval_name}")
        for metric, value in performance.items():
            print(f"{metric}: {value:.2f}")
        print("="*40)

# Convert performance metrics into a DataFrame
df_performance = metrics_to_dataframe(performance_metrics)

# Format the DataFrame for cleaner output
df_performance = df_performance.round(4)
df_performance['Total Return'] = df_performance['Total Return'] * 100  # Convert to percentage

# Display the DataFrame
print(df_performance)
```
